```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('src/OTC_GDP.csv')

sns.set_theme(style="darkgrid")
plt.figure(figsize=(11, 7))

plt.bar(data['Year'] - 0.2, data['OTC Derivatives (USD: trillions)'], width=0.4, label='OTC Derivatives (USD: trillions)', alpha=0.7)
plt.bar(data['Year'] + 0.2, data['Global GDP (USD: trillions)'], width=0.4, label='Global GDP (USD: trillions)', alpha=0.7)

plt.title("OTC Derivatives Relative to Global GDP (2007-2009)", fontsize=24)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Value (USD: trillions)", fontsize=12)
plt.legend(loc="upper left", fontsize=10)

# setting y-axis ticks to intervals of 100, up to 800
# dataset goes up to 700, so 800 is a good upper limit for the legend to be visible
plt.yticks(range(0, 801, 100))

# set the x-axis ticks manually
plt.xticks(data['Year'])
plt.tight_layout()

plt.savefig("img/OTC_GDP.png")
```
