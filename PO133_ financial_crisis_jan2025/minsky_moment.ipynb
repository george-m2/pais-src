```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="dark")

x = np.linspace(0, 10, 200)
y1 = -0.2 * (x - 5)**3 + 2 * (x - 5) + 5 # credit curve
y2 = -0.1 * (x - 5)**3 + 1.5 * (x - 5) + 4 # GDP curve

plt.figure(figsize=(12, 7))

plt.plot(x, y1, 'r-', linewidth=2.5, label='Credit') # red line for credit
plt.plot(x, y2, 'c--', linewidth=2.5, label='GDP') # dashed cyan line for GDP

# intersection points for phases and both lines
plt.scatter(3.5, y1[np.argmin(np.abs(x - 3.5))], color='black', s=50, zorder=5) # y intersection calculated by finding the closest x value
plt.text(3.6, 4.75, 'Speculative\nBorrowing', fontsize=16, color='black')

plt.scatter(6.5, y1[np.argmin(np.abs(x - 6.5))], color='black', s=50, zorder=5) # zorder to force render on top of lines
plt.text(6.6, 8.5, 'Ponzi\nBorrowing', fontsize=16, color='black')

plt.scatter(7.5, y1[np.argmin(np.abs(x - 7.5))], color='black', s=50, zorder=5)
plt.text(7.8, 7, 'Minsky\nMoment', fontsize=16, color='black')

# phase labels
plt.text(1.5, -2, 'Initial economic conditions', ha='center', fontsize=18, fontweight='bold', color='black') #forcing centering
plt.text(5, -2, 'Credit build up phase', ha='center', fontsize=18, fontweight='bold', color='black')
plt.text(8, -2, 'Crisis phase', ha='center', fontsize=18, fontweight='bold', color='black')

# phase lines
plt.axvline(x=3.5, color='black', linestyle='--', linewidth=1)
plt.axvline(x=6.5, color='black', linestyle='--', linewidth=1)

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right', fontsize=18)

# no labels, no ticks
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])

plt.title('Three phases of the Minsky Financial Hypothesis relative to GDP', fontsize=22, color='black')

plt.tight_layout()

plt.savefig("minsky_financial_hypothesis.png")
```
