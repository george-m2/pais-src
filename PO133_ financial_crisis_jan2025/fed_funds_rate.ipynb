```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns

data = pd.read_csv('src/FEDFUNDS.csv')
data['observation_date'] = pd.to_datetime(data['observation_date'])
data.set_index('observation_date', inplace=True)

# seaborn theme, same as marxian rate of profit plot
sns.set_theme(style="darkgrid", font_scale=1.2)
plt.figure(figsize=(11, 7))

plt.plot(data.index, data['FEDFUNDS'], 'o-', alpha=0.75, label='Federal Funds Rate', 
         color='royalblue', markersize=3)

highlight_start = pd.to_datetime('2001-01-01')
highlight_end = pd.to_datetime('2008-09-15')
plt.axvspan(highlight_start, highlight_end, color='yellow', alpha=0.2, 
            label='Build-up to the GFC') #highlight GFC time series

plt.suptitle('Federal Funds Rate (1999-2009)', fontsize=24, y=0.95, x=0.525)
plt.title('Monthly, Not Seasonally Adjusted', fontsize=14)

plt.xlabel('Date')
plt.ylabel('Federal Funds Rate (%)')

y_min, y_max = -2, 8
plt.ylim(y_min, y_max)
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True)) #force integer ticks to avoid incorrect auto datetime ticks

plt.legend(loc='lower left')
plt.tight_layout()

plt.savefig('federal_funds_rate.png', format='png')
```
