```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#https://fred.stlouisfed.org/series/TOTBKCR#0
data_fredgraph = pd.read_csv("src/TOTBKCR_GDP.csv")
# convert the observation_date column to datetime format
data_fredgraph["observation_date"] = pd.to_datetime(data_fredgraph["observation_date"])
# FRED data is in absolute terms (USD), so calculate the growth rate
data_fredgraph["GDP_growth_rate"] = data_fredgraph["GDP"].pct_change() * 100
data_fredgraph["TOTBKCR_growth_rate"] = data_fredgraph["TOTBKCR"].pct_change() * 100

sns.set_theme(style="darkgrid")
plt.figure(figsize=(11, 7))
plt.plot(
    data_fredgraph["observation_date"],
    data_fredgraph["GDP_growth_rate"],
    label="GDP Growth Rate",
    marker="o",
    markersize=4, 
    linewidth=2,   
    color="lightblue"   
)
plt.plot(
    data_fredgraph["observation_date"],
    data_fredgraph["TOTBKCR_growth_rate"],
    label="Credit Growth Rate",
    marker="s",
    markersize=4,
    linewidth=2,
    color="darkorange"
)

plt.title("GDP Growth (%) Relative to Credit Growth (%) (Q4:2000-Q4:2010)", fontsize=20)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Growth Rate (%)", fontsize=12)
plt.legend(loc="lower left")
plt.tight_layout()
plt.savefig("img/TOTBKCR_GDP.png")
```
