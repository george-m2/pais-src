```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
years = ["1992", "1997"]
categories = ["Con (%)", "Lab (%)", "Lib Dem (%)"]
not_centre = {
    "1992": [47, 36, 10],
    "1997": [32, 47, 11],
}
centre = {
    "1992": [40, 32, 19],
    "1997": [22, 49, 19],
}

x = np.arange(len(categories))
width = 0.35

# sharey to make the y-axis the same
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

for i, year in enumerate(years):
    # bars
    rects1 = axs[i].bar(x - width/2, not_centre[year], width, label='Not Centre', color='skyblue')
    rects2 = axs[i].bar(x + width/2, centre[year], width, label='Centre', color='orange')
    
    # percentage labels
    # get the height, width and x position of the bar
    for rect in rects1:
        height = rect.get_height()
        axs[i].text(rect.get_x() + rect.get_width()/2., height,
                   f'{height:.1f}%', ha='center', va='bottom')
    
    for rect in rects2:
        height = rect.get_height()
        axs[i].text(rect.get_x() + rect.get_width()/2., height,
                   f'{height:.1f}%', ha='center', va='bottom')
    
    axs[i].set_title(f"{year}")
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(categories)
    axs[i].set_ylabel('Percentage (%)')
    axs[i].legend()

plt.suptitle("BES 1964-97 Survey on Political Positioning (1992 vs. 1997)", fontsize=15)
plt.tight_layout()
plt.savefig("img/centrist_affiliations.png")
```
