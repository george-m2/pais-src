```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graphs
categories_a = ["Great difference", "Some difference", "No difference"]
data_a_1992 = [56.1, 31.7, 12.2]
data_a_1997 = [33.3, 43.1, 23.6]

categories_b = ["Agree strongly", "Agree", "Neutral", "Disagree", "Disagree strongly"]
data_b_1992 = [10.1, 34.9, 22.0, 31.0, 2.0]
data_b_1997 = [15.5, 42.8, 20.2, 19.9, 1.6]

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# (a) Difference Between Conservatives and Labour
x_a = np.arange(len(categories_a))
axs[0].bar(x_a - 0.2, data_a_1992, 0.4, label='1992', color='skyblue')
axs[0].bar(x_a + 0.2, data_a_1997, 0.4, label='1997', color='orange')
axs[0].set_xticks(x_a)
axs[0].set_xticklabels(categories_a)
axs[0].set_title("""(a) "How much difference Between Conservatives and Labour?" """)
axs[0].set_ylabel("Percentage (%)")
axs[0].legend()

# (b) Opinion on Government Representation
x_b = np.arange(len(categories_b))
axs[1].bar(x_b - 0.2, data_b_1992, 0.4, label='1992', color='skyblue')
axs[1].bar(x_b + 0.2, data_b_1997, 0.4, label='1997', color='orange')
axs[1].set_xticks(x_b)
axs[1].set_xticklabels(categories_b)
axs[1].set_title(""" (b) "People Like Me Have No Say in What Government Does" """)
axs[1].set_ylabel("Percentage (%)")
axs[1].legend()

plt.tight_layout()
plt.savefig("img/party_difference_and_voter_representation.png")
```
