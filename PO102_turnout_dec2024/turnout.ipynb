```python
import matplotlib.pyplot as plt
import numpy as np

years = [2001, 2005, 2010, 2015, 2017, 2019]
turnouts = [59.4, 61.4, 65.1, 66.2, 68.8, 67.3]

# even spacing
x_pos = np.arange(len(years))

plt.figure(figsize=(10, 6))
plt.bar(x_pos, turnouts, width=0.6, color='skyblue', edgecolor='black')
plt.title('UK General Election Turnout (2001-2019)', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Turnout (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7, axis='y')

# year ticks
plt.xticks(x_pos, years)

# % on bars
for x, y in zip(x_pos, turnouts):
    plt.annotate(f'{y}%', (x, y), xytext=(0, 3),
                textcoords='offset points', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig("img/turnout_2001-2019.png")
```
